# 🧠 Dry Bean Classification with TensorFlow

This notebook implements a feed-forward neural network using **TensorFlow** to classify dry bean types. The steps include:

1. Data loading and preprocessing  
2. Model creation and training  
3. Evaluation and interpretation


## 📥 Step 1: Import Libraries and Load Data

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```
# Load dataset
df = pd.read_csv("Lecture_10_Dry_Bean_Dataset.csv")
df.head()
```

## 🔍 Step 2: Data Preprocessing

```
# Encode labels
label_encoder = LabelEncoder()
df['Class_encoded'] = label_encoder.fit_transform(df['Class'])

X = df.drop(columns=['Class', 'Class_encoded'])
y = df['Class_encoded']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# One-hot encode labels for TensorFlow
y_categorical = to_categorical(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_categorical, test_size=0.2, random_state=42, stratify=y)
```

## 🧠 Step 3: Build and Train the Neural Network

```
model = Sequential([
    Dense(64, activation='relu', input_shape=(16,)),
    Dense(32, activation='relu'),
    Dense(7, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.1)
```

## 📈 Step 4: Evaluate Model

```
# Predict
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# Classification report
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_))
```

```
# Confusion matrix
conf_mat = confusion_matrix(y_true, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='d', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## 📌 Step 5: Training Interpretation

- The model was trained using ReLU activations and the Adam optimizer.
- The use of categorical crossentropy is standard for multi-class classification problems.
- The accuracy and loss plots from `model.fit()` provide insight into learning dynamics.

## 📌 Step 6: Evaluation Interpretation

#### 🔹 **Performance Summary**:
- Classes like **BOMBAY**, **CALI**, **HOROZ**, and **SEKER** show strong metrics.
- **SIRA** is more challenging due to overlap with similar classes.

#### 🔹 **Confusion Matrix Insights**:
- Most predictions lie along the diagonal, indicating high accuracy.
- The matrix highlights where misclassifications occur and helps identify confusion between specific bean types.
